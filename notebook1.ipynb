```python
from urllib.request import urlopen
import pandas as pd
from datetime import datetime
import glob




def getcsv(n):
	url="https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_provinceData.php?country=UKR&provinceID=" + str(n) +"&year1=1981&year2=2019&type=Mean";
	vhi_url = urlopen(url)
	now = datetime.now()
	date_time = now.strftime("%m.%d.%Y-%H:%M:%S")
	fileout='vhi_id_' +str(n)+'_' + date_time+'.csv';
	out = open(fileout,'wb')
	out.write(vhi_url.read())
	out.close()
	df = pd.read_csv(fileout, header=1, skipfooter=1)
	print(df)


def todf(path):
	filenames = glob.glob(path+"vhi_id_*.csv")
	dfs = [pd.read_csv(f, header=1, skipfooter=1) for f in filenames]
	data = pd.concat(dfs,ignore_index=True)
	col_list = ['year', 'dfg', 'fdg', 'fdgdf', 'dfgg']
	data.columns = col_list
	print(data)


getcsv(1)
todf('/home/lemonade/python/')
print ("VHI is downloaded...");
```
